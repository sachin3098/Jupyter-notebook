### For single value

```python
import matplotlib.pyplot as plt
%matplotlib inline

blood_sugar=[113,85,90,150,149,88,93,115,80,75,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.95)
```

### For multiple value

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men=[113, 85, 90, 150, 149, 88, 93, 115, 80, 87, 82, 129]
blood_sugar_women=[67, 98, 89, 120, 113, 150, 69, 84, 89, 79, 112, 100]
```

```python
plt.xlabel('Sugar range')
plt.ylabel('Total no.of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'])
plt.legend(shadow=True)
```

```python
plt.xlabel('Sugar range')
plt.ylabel('Total no.of patients')
plt.title('Blood sugar analysis')
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['men','women'],orientation='horizontal')
plt.legend(shadow=True)
```

### Pie chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

### To draw pie chart as perfect circle

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

### To show percentage to everypie and specify the radius.

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.2f%%',radius=1.4)
plt.show()
```

### Now to explode any part of pie element

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.2f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
#To show piechart counterclock or at angle of 180.
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=180)
plt.show()
```

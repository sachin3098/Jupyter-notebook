```python
import matplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,41,40,44,48,47]
avg_t=[45,48,49,42,43,45,49]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

### Now we want to label x-axis and y-axis.

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

### Add Legend into Graph

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')
```

### Bar chart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Simple barchart for showing revenus of Tech companies

```python
company=['GOOGLE','AMAZN','MSFT','FB']
revenue=[90,135,76,82]
```

### We want to show the company name on X-axis, so we convert the list of name into array.

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('Tech comapny Stocks')
plt.legend()
```

### Multiple Bars for showing revenue and profit

```python
profit=[40,45,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

### barh use to show horizontal graph

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

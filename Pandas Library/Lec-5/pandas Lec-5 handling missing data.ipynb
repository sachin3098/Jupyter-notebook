```python
import pandas as pd
df=pd.read_csv("weather_book.csv")
df
```

### Set Date as index

```python
df=pd.read_csv("weather_book.csv",parse_dates=["Day"])
df.set_index("Day",inplace=True)
df
```

### Replace NaN with zero in Entire Table

```python
new_df=df.fillna(0)
new_df
```

### By column wise you can fill Nan with zero

```python
new_df=df.fillna({
    'Temperature':0,
    'Windspeed':0,
    'Event':'no event'
})
new_df
```

### Carry forward previous value if next row is zero.

```python
#In previous table on 1/5/2022 we have temparature 32, on 02/5/2022 we have missing value,so we forward the value.
```

```python
new_df=df.fillna(method="ffill")
new_df
```

### Fill backward value

```python
new_df=df.fillna(method="bfill")
new_df
```

### Use Fillforward value one Time

```python
new_df=df.fillna(method="ffill",limit=1)
new_df
```

### Interpolate method take mean value of previous cell and next cell and fill current empty cell

```python
new_df=df.interpolate()
new_df
```

### In this command while fillling average value it also consider nearest Date also, This Method is more accurate for filling empty cell.

```python
new_df=df.interpolate(method="time")
new_df
```

### To Drop empty Row, if Row has one empty cell,it will drop it.

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how="all")
new_df
```

### Different way of creating DataFrame

### Using read_csv method

```python
import pandas as pd
df=pd.read_csv("weatherbook.csv")
df
```

### Using read excel sheet.

```python
df=pd.read_excel("weatherbook.xlsx","Sheet1")
df
```

### Using python dictionary

```python
import pandas as pd
weatherdata={
            "Day":["1/5/2022","2/5/2022","3/5/2022","4/5/2022","5/5/2022","6/5/2022"],
            "Temperature":[32,35,40,28,42,45],
            "Windspeed":[6,7,5,4,3,2],
            "Event":["Rain","Dusty","Rain","Snow","Sunny","Sunny"]
}
df=pd.DataFrame(weatherdata)
df
```

### Using Tuplelist Datframe method

```python
import pandas as pd
Weather_data=[
    ('1/5/2022','32','6','Rain'),
    ('2/5/2022','35','6','Dusty'),
    ('3/5/2022','40','5','Dusty'),
]
df=pd.DataFrame(Weather_data,columns=["Day","Temperature","Windspeed","Event"])
df
```

### Pivot table allows you to reshape or transfrom Data.

```python
import numpy as np
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/10_pivot/weather.csv")
df
```

### Pivot Basic


```python
#we create two columns and reshape our data in Two columns.
```

```python
df.pivot(index="city",columns="date")
```

### Print only Humidity value

```python
df.pivot(index="city",columns="date",values="humidity")
```

### Pivot Table

```python
# Here we have Two vlaue for Temperature and humidity on same Date, so when create pivot table it automatically take mean vlaue.
```

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/10_pivot/weather2.csv")
df
```

```python
#Mean value of Tempearture and Humidity
```

```python
df.pivot_table(index="city",columns="date")
```

### Margins

```python
# Here we add value in rows and columns wise.
```

```python
df.pivot_table(index="city",columns="date",margins=True, aggfunc=np.sum)
```

### Grouper Function

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/10_pivot/weather3.csv")
df
```

```python
# Now we want to show mean value at the end of month,so we use below command.
#Here freq=M Which is Month.
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weatherbook1.csv")
df
```

### Replacing List with single value

```python
new_df = df.replace(to_replace=[-9999,-8888], value=0)
new_df
```

### Replacing per column

```python
new_df=df.replace({
    'Temperature':-9999,
    'Windspeed':-9999,
    'Event':'0'
},np.NaN)
new_df
```

### Regex expression

```python
new_df=df.replace({
    'Temperature':'[A-Za-z]',
    'Windspeed':'[A-Za-z]',
    },'',regex=True)
new_df
```

### Replacing List with another List

```python
df=pd.DataFrame({
    'score':['exceptional','average','good','poor','average','exceptional'],
    'student':['rob','maya','parthiv','tom','julian','erica']
})
df
```

```python
new_df=df.replace(['poor','average','good','exceptional'],[1,2,3,4])
new_df
```
